```python
#task2
#Read frames from any camera. Press Q to terminate the code,
# R to rotate the frame 90 degrees,
# C to save the frame on your device,
# S to save a video stream using OpenCV.
# G to convert the frame to grayscale,
# H to convert the frame to HSV space,
# and X to show (grayscale, HSV, rotated and original frames) together, Z to show only the original frame.


import cv2

cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    cv2.imshow('frame', frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    if cv2.waitKey(1) & 0xFF == ord('r'):
        image = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        cv2.imshow('rotate',image)

    if cv2.waitKey(1) & 0xFF == ord('g'):
        gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow('gray',gray)

    if cv2.waitKey(1) & 0xFF == ord('h'):
        hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        cv2.imshow('HSV',hsv)

    if cv2.waitKey(1) & 0xFF == ord('c'):
        cv2.imwrite('vortex.jpg', frame)

    if cv2.waitKey(1) & 0xFF == ord('s'):
        fourcc = cv2.VideoWriter_fourcc(*'XVID')
        out = cv2.VideoWriter('output.avi', fourcc, 10, (1024, 720))

    if cv2.waitKey(1) & 0xFF == ord('x'):
        image = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        cv2.imshow('rotate', image)
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        cv2.imshow('gray', gray)
        hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
        cv2.imshow('HSV', hsv)

    if cv2.waitKey(1) & 0xFF == ord('z'):
        cv2.destroyAllWindows


cap.release()
cv2.destroyAllWindows()

cap = cv2.VideoCapture(0)

```

```python
#task4
import cv2 as cv
import numpy as np

circles = []
img = np.zeros((512, 512, 3), np.uint8)


# mouse click
def mousePoints(event, x, y, flags, params):
    if event == cv.EVENT_LBUTTONDOWN:
        circles.append([x, y])  # add circle
        print("EVENT_LBUTTONDOWN")

    if event == cv.EVENT_RBUTTONDOWN:
        circles.pop()  # remove circle
        print("EVENT_RBUTTONDOWN")


while 1:
    img=np.zeros((512,512,3),np.uint8)
    for i in circles:

        cv.circle(img,i,75,(50,300,240),3) #img,center,rad,color,thickness
    cv.imshow("Circles", img)
    cv.setMouseCallback("Circles", mousePoints)
    if cv.waitKey(1) & 0xFF== ord('q'):
           break
```
